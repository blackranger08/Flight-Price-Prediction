```python
#Importing Libraries
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementNotInteractableException
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
import os
import urllib
import requests
```

```python
name=[]
dt=[]
at=[]
dc=[]
ac=[]
ttt=[]
ns=[]
fare=[]
#importing webdriver
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()  #maximizing window
time.sleep(1)
urls="https://flight.yatra.com/air-search-ui/seodom/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=DEL&destinationCountry=IN&flexi=0&flight_depart_date=23%2F11%2F2021&hb=&noOfSegments=1&origin=CCU&originCountry=IN&type=O&unique=1634237229043&viewName=normal"
driver.get(urls) #loading url
time.sleep(45)


    
#name

try:
    c=driver.find_elements_by_xpath("//span[@class='i-b text ellipsis']")
    for i in c:
        name.append(i.text)
    
except NoSuchElementException:
    name.append("--")

#Departure time
try:
    d=driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']")
    for i in d:
        dt.append(i.text)
    
except NoSuchElementException:
    dt.append("--")
        
#Arrival Time
try:
    e=driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']")
    for i in e:
        at.append(i.text)
    
except NoSuchElementException:
     at.append("--")
        

#Departure city
try:
    f=driver.find_elements_by_xpath("//p[@class='fs-10 font-lightgrey no-wrap city ellipsis']")
    for i in f:
        dc.append(i.text)
    
except NoSuchElementException:
    dc.append("--")

    
#Arrival city
try:
    g=driver.find_elements_by_xpath("//p[@class='fs-10 font-lightgrey no-wrap city ellipsis']")
    for i in g:
        ac.append(i.text)
    
except NoSuchElementException:
    ac.append("--")
    
#Total Time taken 
try:
    h=driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']")
    for i in h:
        ttt.append(i.text)
    
except NoSuchElementException:
    ttt.append("--")
    
#NUmber of stops
try:
    q=driver.find_elements_by_xpath("//div[@class='stop-cont pl-13']")
    for i in q:
        ns.append(i.text)
    
except NoSuchElementException:
    ns.append("--")

#Fare
try:
    j=driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")
    for i in j:
        fare.append(i.text)
    
except NoSuchElementException:
    fare.append("--")




dc=dc[::2]
ac=ac[1::2]
ans10=pd.DataFrame({})
ans10["Plane Name"]=name
ans10["Departure Time"]=dt
ans10["Arrival Time"]=at
ans10["Departure City"]=dc
ans10["Arrival City"]=ac
ans10["Total Time Taken"]=ttt
ans10["Numbers of stops"]=ns
ans10["Fare"]=fare
ans10

        
        
    
    
```
